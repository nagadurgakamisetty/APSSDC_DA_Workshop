### robust Scalar
-  xi-Q2(x)
---------------------
- Q3(x)-Q(1)

#### inter Quatile range(IQR)
- Q1->25%(Lower Quartile)
- Q2->50%(medium quartile)
- Q3->75%(upper quartile)

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
data1
```

```python
data1.columns
```

```python
data1.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data1.head(20)
```

### apply robust scalar model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs=RobustScaler()
rs
```

```python
rs2=rs.fit_transform(data1)
rs2
```

```python
rs2=pd.DataFrame(rs2,columns=data1.columns)
rs2.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(9,7))
sns.kdeplot(data1['TV'],ax=ax1)
sns.kdeplot(data1['radio'],ax=ax1)
sns.kdeplot(data1['newspaper'],ax=ax1)
sns.kdeplot(data1['sales'],ax=ax1)

sns.kdeplot(rs2['TV'],ax=ax2)
sns.kdeplot(rs2['radio'],ax=ax2)
sns.kdeplot(rs2['newspaper'],ax=ax2)
sns.kdeplot(rs2['sales'],ax=ax2)
```
